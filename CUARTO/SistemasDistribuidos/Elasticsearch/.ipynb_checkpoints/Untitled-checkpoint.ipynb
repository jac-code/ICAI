```python
import utm

def convert_from_utm(easting, northing, zone, letter):
    return utm.to_latlon(easting, northing, zone, letter)
```

```python
import unidecode 

def convert_to_utf8(cadena):
    return unidecode(cadena.replace("Ø", "n.").replace("é","e").replace("á","a").replace("ú","u").replace("ó","o").replace("í","i").replace("'","").replace("ñ","ni"))
```

```python
import requests
import json
from xml.etree import ElementTree
from datetime import datetime

url = 'https://practica-elastic-search.es.us-central1.gcp.cloud.es.io:9243/'
username = 'superuser'
password = 'admin1234'
index = 'trafico-madrid'

headers = {"Content-Type": "application/json"}

url_fichero = "https://informo.madrid.es/informo/tmadrid/pm.xml"

response = requests.get(url_fichero)
root = ElementTree.fromstring(response.content)

# en la variable out tenemos los datos del fichero
out = []
for pm in root[1:-1]:
    d = {}
    for i in pm:
        d[i.tag] = i.text
    
    out.append(d)
print("Hay ", len(out), " medidas en el XML")

for i in out:
    try:
        [dia, hora] = str(datetime.strptime(root[0].text, "%d/%m/%Y %H:%M:%S")).split(" ")
        i["fecha_hora"] = dia + "T" + hora + "Z"
    except KeyError:
        pass
    try:
        i["idelem"] = convert_to_utf8(i["idelem"])
    except KeyError:
        pass
    except:
        del i["idelem"]
    try:
        i["descripcion"] = convert_to_utf8(i["descripcion"])
    except KeyError:
        pass
    except:
        del i["descripcion"]
    try:
        i["accesoAscociado"] = convert_to_utf8(i["accesoAscociado"])
    except KeyError:
        pass
    except:
        del i["accesoAscociado"]
    try:
        i["intensidad"] = int(i["intensidad"])
    except KeyError:
        pass
    except:
        del i["intensidad"]
    try:
        i["ocupacion"] = int(i["ocupacion"])
    except KeyError:
        pass
    except:
        del i["ocupacion"]
    try:
        i["carga"] = int(i["carga"])
    except KeyError:
        pass
    except:
        del i["carga"]
    try:
        i["nivelServicio"] = int(i["nivelServicio"])
    except KeyError:
        pass
    except:
        del i["nivelServicio"]
    try:
        i["intensidadSat"] = int(i["intensidadSat"])
    except KeyError:
        pass
    except:
        del i["intensidadSat"]
    try:
        i["error"] = convert_to_utf8(i["error"])
    except KeyError:
        pass
    except:
        del i["error"]
    try:
        i["subarea"] = convert_to_utf8(i["subarea"])
    except KeyError:
        pass
    except:
        del i["subarea"]
    try:
        x = i["st_x"]
        y = i["st_y"]
        x = x.replace(',', '.')
        x = float(x)
        y = y.replace(',', '.')
        y = float(y)

        [lat, lon] = convert_from_utm(x, y, 30, 'T')
        
        del i["st_x"]
        del i["st_y"]
        i["coordenadas"] = {"lat": lat, "lon": lon}
        
    except KeyError:
        pass
    except:
        del i["coordenadas"]
    
    data = str(i).replace("'", "\"")
    response = requests.post(url+index+"/_doc", data=data, headers=headers, auth=(username, password))
```

```python
import requests
import json
from xml.etree import ElementTree
from datetime import datetime

url = 'https://practica-elastic-search.es.us-central1.gcp.cloud.es.io:9243/'
username = 'superuser'
password = 'admin1234'
index = 'trafico-madrid'

headers = {"Content-Type": "application/json"}

url_fichero = "https://informo.madrid.es/informo/tmadrid/pm.xml"

response = requests.get(url_fichero)
root = ElementTree.fromstring(response.content)

# en la variable out tenemos los datos del fichero
out = []
for pm in root[1:-1]:
    d = {}
    for i in pm:
        d[i.tag] = i.text
    
    out.append(d)
print("Hay ", len(out), " medidas en el XML")

ndjson = ""
for i in out:
    try:
        [dia, hora] = str(datetime.strptime(root[0].text, "%d/%m/%Y %H:%M:%S")).split(" ")
        i["fecha_hora"] = dia + "T" + hora + "Z"
    except KeyError:
        pass
    try:
        i["idelem"] = convert_to_utf8(i["idelem"])
    except KeyError:
        pass
    except:
        del i["idelem"]
    try:
        i["descripcion"] = convert_to_utf8(i["descripcion"])
    except KeyError:
        pass
    except:
        del i["descripcion"]
    try:
        i["accesoAsociado"] = convert_to_utf8(i["accesoAsociado"])
    except KeyError:
        pass
    except:
        del i["accesoAsociado"]
    try:
        i["intensidad"] = int(i["intensidad"])
    except KeyError:
        pass
    except:
        del i["intensidad"]
    try:
        i["ocupacion"] = int(i["ocupacion"])
    except KeyError:
        pass
    except:
        del i["ocupacion"]
    try:
        i["carga"] = int(i["carga"])
    except KeyError:
        pass
    except:
        del i["carga"]
    try:
        i["nivelServicio"] = int(i["nivelServicio"])
    except KeyError:
        pass
    except:
        del i["nivelServicio"]
    try:
        i["intensidadSat"] = int(i["intensidadSat"])
    except KeyError:
        pass
    except:
        del i["intensidadSat"]
    try:
        i["error"] = convert_to_utf8(i["error"])
    except KeyError:
        pass
    except:
        del i["error"]
    try:
        i["subarea"] = convert_to_utf8(i["subarea"])
    except KeyError:
        pass
    except:
        del i["subarea"]
    try:
        x = i["st_x"]
        y = i["st_y"]
        x = x.replace(',', '.')
        x = float(x)
        y = y.replace(',', '.')
        y = float(y)

        [lat, lon] = convert_from_utm(x, y, 30, 'T')
        
        del i["st_x"]
        del i["st_y"]
        i["coordenadas"] = {"lat": lat, "lon": lon}
        
    except KeyError:
        pass
    except:
        del i["coordenadas"]
    
    
    data = str(i).replace("'", "\"")
    ndjson += """{"index":{"_index": "trafico-madrid"}}""" + "\n" + data + "\n"
        
response = requests.post(url+index+"/_bulk", data=ndjson, headers=headers, auth=(username, password))
response.json()
```
