```python
import pickle
from datetime import date
from datetime import datetime
from socket import *

s = socket(AF_INET, SOCK_STREAM)

s.bind(("127.0.0.1", 6780))
s.listen(1)
while True:
    #(conn, addr) = s.accept()
    data = conn.recv(1024)
    msg = pickle.loads(data)
    print('El cliente ' + str(msg[1]) + ' ha solicitado la hora y la fecha.')
    
    now = datetime.now()
    current_time = now.strftime("%H:%M:%S")
    today = date.today()

    msg_servidor_saludo = 'Hola cliente ' + str(msg[1]) + ', te envío la información solicitada.'
    msg_servidor_hora = 'La hora es: ' + str(current_time)
    msg_servidor_fecha = 'La fecha es: ' + str(today)
    bytes_tx = pickle.dumps([msg_servidor_saludo, msg_servidor_hora, msg_servidor_fecha, current_time, today])
    conn.send(bytes_tx)
    print('Mensaje enviado.')
    conn.close()
```
