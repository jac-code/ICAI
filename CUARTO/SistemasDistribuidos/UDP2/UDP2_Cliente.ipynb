APARTADO 1

Este cliente, mediante UDP, etablece una conexión con un servidor situado en la dirección server_address. Le manda el mensaje y su identificador. En este caso el cliente tratará de enviar el mensaje aunque el servidor no esté levantado, si pasados 5 segundos no consigue realizar la conexión, desistriá y cerrara el socket. Si recibe la información, mostrará la hora y la fecha.

```python
import time
import pickle
import socket

id_cliente = 1

msg = ''
cont = 0
msg = '¿Me puedes mandar la hora?'

server_address = ("127.0.0.1", 6780)
bytes_tx = pickle.dumps((msg, id_cliente))

s = socket.socket(family=socket.AF_INET, type=socket.SOCK_DGRAM)
for cont in range(0,5):
    try:
        s.sendto(bytes_tx,server_address)
        bytes_rx = s.recvfrom(1024)
        mensaje_rec = pickle.loads(bytes_rx[0])
        hora = str(mensaje_rec[1])
        fecha = str(mensaje_rec[2])
        break
    except:
        cont = cont + 1
        print('Esperando...')
        time.sleep(1)
        if cont == 10:    
            print('Conexión cerrada')
            break
if (hora != ''):
    print(hora)
    print(fecha)
    print('Message received from server: ' +str(bytes_rx[1]))

s.close()
```
