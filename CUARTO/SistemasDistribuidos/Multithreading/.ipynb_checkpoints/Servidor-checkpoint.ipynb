# SERVIDOR MULTITHREADING

APARTADO 1

Este es un servidor, que en primer lugar, pide el puerto en el que levantarse. Como el puerto tiene que coincidir con al que se conecta el cliente, le especificamos el puerto que tiene indicado el cliente. El servidor recibe los mensajes de los clientes que son manejados cada uno en un hilo diferente. Finalmete el servidor reenvia el mensaje al cliente y cierra esa conexión.

```python
import socket
import sys
import traceback
from threading import Thread

def main():
     start_server()
        
def start_server():
    host = "localhost"
    port = 0
    
    while(port == 0):
        try:
            print("Introduzca el número del puerto (6780 para el servidor): ")
            port = int(input())
            print(port)
        except:
            print('El puerto debe ser un numero')
    
    sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    sock.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR, 1)
    print("Socket created")
    try:
        sock.bind((host, port))
    except:
        print("Bind failed. Error : " + str(sys.exc_info()))
        sys.exit()
    sock.listen() 
    # infinite loop- do not reset for every requests
    while True:
        print("Socket now listening")
        connection, address = sock.accept()
        ip, port = str(address[0]), str(address[1])
        print("Connection from client " + ip + ":" + port)
        try:
            Thread(target=clientThread, args=(connection, ip, port)).start()
        except:
            print("Thread did not start.")
            traceback.print_exc()
    sock.close()
    
def clientThread(connection, ip, port, max_buffer_size = 1024):
    is_active = True
    cont = 0
    while is_active:
        msg = connection.recv(max_buffer_size).decode()
        print(msg)
        cont = cont + 1
        if msg == 'exit':
            print("Client {} is requesting to quit".format(cont))
            connection.close()
            print("Connection " + ip + ":" + port + " closed")
            is_active = False
        else:
            print("Client {} sent data: {}".format(cont,msg))
            connection.send("ok".encode())

if __name__ == "__main__":
    main()
```

Como podemos ver en ocasiones tenemos mensajes incompletos o mensajes que se mezclan entre sí. Esto es debido a que a pesar de que utilizamos hebras diferentes para cada uno de los mensajes, tienen recursos compartidos entre ellas. En este caso, el más evidente es el comando de imprimir por pantalla, es por eso que si no manejamos con precaución el recurso de la pantalla, nos pueden mostrar mensajes confundidos

APARTADO 2

En este caso el servidor es igual que el del apartado anterior pero siendo capaces de manejar correctamente los recursos compartidos por las hebras. Haremos uso del comando threading.Lock(). Esto nos permite que mientras una hebra esté imprimiendo por pantalla, otra pueda interrumpir el proceso. Utilizaremos este comando en cada uno de los print a los que el uso de hebras diferentes pueda molestar.

```python
import socket
import sys
import traceback
import threading

global threadLock 
threadLock = threading.Lock()

def main():
     start_server()
        
def start_server():
    host = "localhost"
    port = 0
    
    while(port == 0):
        try:
            print("Introduzca el número del puerto (6780 para el servidor): ")
            port = int(input())
            print(port)
        except:
            print('El puerto debe ser un numero')
    
    sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    sock.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR, 1)
    print("Socket created")
    try:
        sock.bind((host, port))
    except:
        print("Bind failed. Error : " + str(sys.exc_info()))
        sys.exit()
    sock.listen() 
    # infinite loop- do not reset for every requests
    while True:
        threadLock.acquire(1)
        print("Socket now listening")
        threadLock.release()
        connection, address = sock.accept()
        ip, port = str(address[0]), str(address[1])
        threadLock.acquire(1)
        print("Connection from client " + ip + ":" + port)
        threadLock.release()
        try:
            threading.Thread(target=clientThread, args=(connection, ip, port)).start()
        except:
            print("Thread did not start.")
            traceback.print_exc()
    sock.close()
    
def clientThread(connection, ip, port, max_buffer_size = 1024):
    is_active = True
    cont = 0
    while is_active:
        msg = connection.recv(max_buffer_size).decode()
        #cont = cont + 1
        if msg == 'exit':
            threadLock.acquire(1)
            print("Client {} is requesting to quit".format(cont))
            threadLock.release()
            connection.send('Conexión terminada.'.encode())
            connection.close()
            #print("Connection " + ip + ":" + port + " closed")
            is_active = False
        else:
            threadLock.acquire(1)
            print("Client sent data: {}".format(msg))
            threadLock.release()
            connection.send("ok".encode())

if __name__ == "__main__":
    main()
```
