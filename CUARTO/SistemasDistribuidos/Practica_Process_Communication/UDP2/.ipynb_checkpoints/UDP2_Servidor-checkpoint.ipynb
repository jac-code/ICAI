# UDP 2 - SERVIDOR

APARTADO 1 y 2 --> sirve para ambos

- Este servidor nos proporciona, al conectarnos desde el cliente, la informacion de la fecha y hora, las veces que los solicitemos. 
- Podremos ver qué cliente nos hace la solicitud y un mensaje de control para saber si la información ha sido enviada correctamente. El servidor quedará levantada hasta que cerremos la conexión manualmente.

```python
import pickle
from datetime import date
from datetime import datetime
import socket

s = socket.socket(family=socket.AF_INET, type=socket.SOCK_DGRAM)

s.bind(("127.0.0.1", 6780))
while True:
    bytes_rx=s.recvfrom(1024)
    msg = pickle.loads(bytes_rx[0])
    print('El cliente ' + str(msg[1]) + ' ha solicitado la hora y la fecha.')
    
    now = datetime.now()
    current_time = now.strftime("%H:%M:%S")
    today = date.today()

    msg_servidor_saludo = 'Hola cliente ' + str(msg[1]) + ', te envío la información solicitada.'
    msg_servidor_hora = 'La hora es: ' + str(current_time)
    msg_servidor_fecha = 'La fecha es: ' + str(today)
    bytes_tx = pickle.dumps((msg_servidor_saludo, msg_servidor_hora, msg_servidor_fecha, current_time, today))
    s.sendto(bytes_tx,bytes_rx[1])
    print('Mensaje enviado.')
```

APARTADO 3

- El servidor mandará una respuesta al cliente cada vez que se establezca la conexión.

```python
import pickle
from datetime import date
from datetime import datetime
import socket

s = socket.socket(family=socket.AF_INET, type=socket.SOCK_DGRAM)
s.bind(("127.0.0.1", 6780))

while True:
    bytes_rx=s.recvfrom(1024)
    msg = pickle.loads(bytes_rx[0])
    print('Mensaje nº: ' + msg + ' recibido')

    
    msg_servidor_confirmacion = 'Paquete recibido'
    bytes_tx = pickle.dumps((msg_servidor_confirmacion))
    s.sendto(bytes_tx,bytes_rx[1])
    
s.close()
```
